```python
# validate the quality and correctness of email users
email = 'lisagamtilon@.com'
# clean the string
email = email.strip()
# email must not be empty
if email == '':
  print('Email cannot be empty')
# email must cotain ',' and one '@' symbol
if not('.' in email and '@' in email):
  print('Email must contain . and @')
# email must contain exactly one '@' symbol:
if email.count('@') != 1:
  print('Email must contain exactly one @')
#email must end with .com, .org, .net
if not(email.endswith(('.com','.org','.net'))):
  print('Email must end with .com, .org, .net')
# email must not be longer than 254 characters:
if len(email) > 254:
  print('Email must not be longer than 254 characters)')
# email must start and end with a letter or digit
if not(email[0].isalnum() and email[-1].isalnum()):
  print('email must start and end with a letter or digit')
else:
  print("Email is valid")
```

```python
# password requirements:
# Must not be empty
# Must be at least 8 characters
# Include at least 1 uppercase letter
# Include at least 1 lowercarse letter
# Not be the same as email
# Not contain any spaces
# Must start and end with a letter or digit
import re
password = 'lisagam tilon@.co()'
password = password.strip()

if password == '':
  print('Password cannot be empty')
if not(len(password) >= 8):
  print('Password must contain at least 8 characters')
if not re.search('[A-Z]', password):
  print('Password must include at least 1 Uppercase letter')
if not re.search('[a-z]',password):
  print('Password must include at least 1 Lowercase letter')
if password == email:
  print('Password cannot be the same as email')
if re.search(r"\s", password):
  print('Password cannot contain spaces')
if not(password[0].isalnum() and password[-1].isalnum()):
  print('Password must start and end with a letter of digit')
else:
  print("Password is valid")
```
