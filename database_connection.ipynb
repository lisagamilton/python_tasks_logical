```python
from google.cloud import bigquery
from google.oauth2 import service_account
import pandas as pd
```

```python
credentials = service_account.Credentials.from_service_account_file('/content/credentials.json')
```

```python
project_id = 'my-project-26-01-2026'
```

```python
client = bigquery.Client(credentials=credentials,project=project_id)
```

```python
query_text = f"""SELECT * FROM `my-project-26-01-2026.dataset1.customers`
LIMIT 100;"""
```

```python
query = client.query(query_text)
```

```python
result_df = query.to_dataframe()
result_df
```

```python
result_df = result_df[result_df['invoice_no'].notnull()]
result_df
```

```python
result_df.info()
```

```python
result_df.isna().count()
```

```python
result_df = result_df.drop_duplicates()
```

```python
#Publishing data to Google Big Query
#it's also important to change the path and project you want to export your data
#so that you won't be modifying original data if something goes wrong - it will be
#irreversible.
result_df.to_gbq(
    destination_table="dataset1.customers", #here we are changing original data
    project_id="my-project-26-01-2026", #saving to the same project
    if_exists="replace", #we are deleting the previous table, we can also append
    credentials=credentials
)
```
