```python
import pandas as pd
clients_q1_df = pd.DataFrame({
    'ClientID':[1,2],
    'ClientName':['Alice','Bob'],
    'Quarter':['Q1','Q1']
})

clients_q2_df = pd.DataFrame({
    'ClientID':[3,4],
    'ClientName':['Charlie','David'],
    'Quarter':['Q2','Q2']
})

concatenated_df = pd.concat([clients_q1_df,clients_q2_df],axis=0)
print(concatenated_df)
```

```python
sales_df = pd.DataFrame({
    'Region':['North','South','North','South','East','East'],
    'Sales':[200,150,300,350,100,250]
})

grouped_df = sales_df.groupby('Region')['Sales'].agg('mean')
grouped_df
```

```python
sales_df = pd.DataFrame({
    'Region':['North','South','North','South','East','East'],
    'Product':['A','A','B','B','A','B'],
    'Sales':[200,150,300,350,100,250]
})

pivot_df = pd.pivot_table(sales_df,values='Sales',index='Region',columns='Product',aggfunc='sum')
pivot_df
```

```python
pivot_df = pd.pivot_table(sales_df,
                          values='Sales',
                          index='Region',
                          columns='Product',
                          aggfunc=['sum','mean','count'],
                          fill_value=0)
pivot_df
```

```python
customers = pd.read_csv('/content/customers.csv')
```

```python
payment = pd.read_csv('/content/payment_method.csv')
```

```python
shopping_data = pd.read_csv('/content/customer_shopping_data.csv')
```

```python
customers.head(5)
```

```python
payment.head(5)
```

```python
shopping_data.head(5)
```

```python
merged_df = pd.merge(shopping_data,customers,on='customer_id',how='left')
```

```python
df = pd.merge(merged_df,payment,left_on='payment_method',right_on='payment_type_id',how='left')
df
```

```python
df.drop(columns=['payment_type_id'],axis=1,inplace=True)
```

```python
df = df.rename({'payment_method_x':'payment_method_id','payment_method_y':'payment_method'}, axis="columns")
```

```python
df.info()
```

```python
df['price'] = df['price'].replace(',','.',regex=True).astype('float')
```

```python
df
```

```python
df.info()
```

```python
df['gender'] = df['gender'].fillna('No_info')
```

```python
# Quantity of clients based on gender
df.groupby('gender')['customer_id'].count()
```

```python
df['total_purchase'] = df['price'] * df['quantity']
```

```python
# Customer count and total purchase value by store and gender
df.groupby(['gender','shopping_mall'])['total_purchase'].agg(['count','sum'])
```

```python
# Total purchase value by payment method
df.groupby('payment_method')['total_purchase'].sum()
```

```python
# Create pivot in which show: 1. Shopping mall 2.  Clients quantity 3.  Total purchase value 4.  Average receipt
pivot_df = pd.pivot_table(df,
                          values =['total_purchase'],
                          index='shopping_mall',
                          aggfunc=['count','sum','mean']
                          )
pivot_df
```

```python
pivot_df.columns = ['count', 'sum', 'mean']
```

```python
pivot_df
```

```python
pivot_df.to_csv('pivot_df.csv', index=True)
```
